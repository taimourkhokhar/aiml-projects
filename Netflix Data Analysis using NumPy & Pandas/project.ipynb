```python
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

install('openpyxl')
install('pandas')
install('matplotlib')
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
df.dropna(subset=['director', 'country'], inplace=True)

df_country = df['country'].value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(df_country, labels=df_country.index, autopct="%1.1f%%")
plt.title("Top 5 Countries")
plt.show()
```
