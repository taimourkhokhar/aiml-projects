```python
#Load and Data Cleaning Process
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
df.dropna(subset=['type','country','release_year'],inplace=True)
df.head(5)
```

```python
from matplotlib import pyplot as plt
type_count=df['type'].value_counts()
plt.figure()
type_count.plot(kind='bar')
plt.title("Movies and Tv show on Netflix")
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
top_countries=df['country'].value_counts().head(5)
plt.figure()
top_countries.plot(kind='pie',autopct='%1.1f%%')
plt.title("Top 5 Movies producing Countries")
plt.ylabel('')
plt.show()
```

```python
year_data=df['release_year'].value_counts().sort_index()
plt.figure()
year_data.plot()
plt.title("Trend over the past years")
plt.xlabel("Years")
plt.ylabel("Number of shows")
plt.show()
```

```python
import seaborn as sns
plt.figure()
sns.countplot(data=df,x='type')
plt.title('Content Distribution Seaborn')
plt.show()
```
