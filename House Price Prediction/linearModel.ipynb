```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset (simple example)
x = np.array([500, 1000, 1500, 2000, 2500])   # house size
y = np.array([100, 200, 300, 400, 500])       # price

def gradient_descent(x, y):
    m = b = 0
    lr = 0.0000001
    iterations = 1000
    n = len(x)

    for i in range(iterations):
        y_pred = m * x + b

        cost = (1/n) * np.sum((y - y_pred) ** 2)

        md = -(2/n) * np.sum(x * (y - y_pred))
        bd = -(2/n) * np.sum(y - y_pred)

        m = m - lr * md
        b = b - lr * bd

        if i % 100 == 0:
            print(f"Iteration {i}: cost={cost}, m={m}, b={b}")

    return m, b

m, b = gradient_descent(x, y)

# Prediction line
y_pred_line = m * x + b

# Plot
plt.scatter(x, y, color="red", label="Actual Data")
plt.plot(x, y_pred_line, color="blue", label="Best Fit Line")
plt.legend()
plt.show()

print("Final equation: y =", m, "* x +", b)
```
